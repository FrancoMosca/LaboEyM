### _Introducción_
La resistencia eléctrica se refiere a la medida de la oposición que presenta un material a la circulación de la corriente eléctrica a través de él. Su unidad de medición es el ohmio ${(\Omega)}$. Esta característica depende de las dimensiones del resistor y del material con el que está construido. Por otro lado, la resistividad ${\rho}$ es una propiedad física que caracteriza la capacidad de un material para oponerse al flujo de la corriente eléctrica y depende directamente del tipo de material, su temperatura y pureza.

La ley de Ohm establece que la corriente I es directamente proporcional al potencial V e inversamente proporcial a la resistencia del conductor R, entonces: $${I = \frac{\Delta V}{R} \Longrightarrow R = \frac{\Delta V}{I}}$$

### _Procedimiento experimental_

```python
import matplotlib.pyplot as plt
import numpy as np

d_values = [1.9e-3, 2.6e-3, 3.6e-3, 4.4e-3, 4.5e-3, 5.2e-3, 6.2e-3, 7.0e-3, 7.8e-3, 8.7e-3, 9.6e-3, 10.5e-3] # Valores de Corriente [A]
C_values = [1.06,  1.46, 2.02, 2.5, 2.51, 2.99, 3.52, 4.0, 4.47, 4.96, 5.49, 6.01] # Valores de Voltaje [V]

fig, ax = plt.subplots()

# Graficar los valores de corriente vs voltaje
ax.plot(d_values, C_values, marker='o', linestyle='-', color='r')

# Ajuste de una línea a los datos
slope, intercept = np.polyfit(d_values, C_values, 1)

# Añadir la ecuación de la línea (pendiente)
equation_text = f'Pendiente: {slope:.4f} V/A'
ax.text(0.005, 5.5, equation_text, fontsize=12, color='blue')

# Etiquetas de los ejes
ax.set_xlabel('Corriente (A)')
ax.set_ylabel('Voltaje (V)')

# Título del gráfico
ax.set_title('Curva Voltaje vs Corriente de la Lamparita (Nuevo Conjunto)')

# Mostrar la cuadrícula
plt.grid(True)

# Mostrar el gráfico
plt.show()
```
