```python
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter

def not_cientifica(numero, cifra=2):
    resultado = f"{numero:.{cifra}e}"
    return resultado

B = 3.47e-4 * 32.2
N = 200
A =  math.pi * (0.015 ** 2)
w = 2 * math.pi * 100
t = 0.1

flujo = B * A * N                           
print(f"flujo magnetico = {not_cientifica(flujo)} * sin(w*t)")

fem_concat = f"fem = {not_cientifica(flujo)} * cos(w*t)"
print(fem_concat)

# Grafico de la fem
amplitud = 1.58e-3      
frecuencia = 100               #frecuencia en Hz 
t = np.linspace(0, 0.10, 1000)  #tiempo

fem = amplitud * np.cos(2 * np.pi * frecuencia * t)

plt.figure(figsize=(10, 5))
plt.plot(t, fem, label="FEM")
plt.xlabel("Tiempo (s)")
plt.ylabel("FEM (V)")
plt.title("Fem en funcion del tiempo")
plt.grid(True)
plt.legend()

# Formateo para notacion cientifica
plt.gca().yaxis.set_major_formatter(ScalarFormatter(useMathText=True))
plt.gca().yaxis.get_major_formatter().set_powerlimits((0, 0))

plt.show()
```
