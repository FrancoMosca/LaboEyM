### _Introducción_
Un capacitor es un dispositivo formado por dos placas o láminas conductoras separadas entre sí por un dieléctrico especialmente diseñado para almacenar una cantidad finita de energía. La unidad de la capacidad es el Faradio ${[F]}$. 

Un dieléctrico es un medio eléctricamente aislante que se opone al paso de la corriente eléctrica.

Si se conecta el capacitor a un generador se produce un pasaje de electrones hasta que las dos placas consiguen albergar el mismo número de cargas pero de _signo contrario_. Una vez cargado, la corriente que circula por el circuito es nula y aunque se desconecte el generador, el capacitor no se descarga. 

La forma más fácil de armar un capacitor es colocar dos placas paralelas enfrentadas entre sí separadas por una distancia que es pequeña en comparación con sus dimensiones. Cuando las placas están cargadas se genera un campo eléctrico ${E}$ uniforme entre medio de las placas. La magnitud de ese campo eléctrico se puede calcular utilizando la Ley de Gauss o el principio de superposición de campos eléctricos, entonces:  ${E = \frac{\sigma}{\epsilon_{0}}}$, donde ${\epsilon_{0}}$ es la permitividad eléctrica en el vacío, con un valor de ${8,85.10^{-12} \frac{F}{m}}$.

Si se hace extensiva esta igualdad a toda la placa del capacitor, el campo eléctrico dentro del mismo queda definido como: ${E = \frac{q}{\epsilon_{0}.A'}}$, donde _q_ es la carga neta de cada una de las placas y _A_ es el área de cada una.

Dado que el campo es uniforme y la distancia entre las placas es _d_, la diferencia de potencial entre ambas será: $${\Delta V_{ab}= E . d = \frac{q}{\epsilon_{0}.A}}$$

Entonces, la capacidad de un capacitor plano en el vacío queda definida como: $${C= \frac{q}{\Delta V_{ab}} = \frac{\epsilon_{0}.A}{d}}$$

La capacitancia varia si, en vez de vacío, hay otro tipo de material entre las placas. Esta se va a ver afectada por las características de este componente particular por lo que su ecuación cambiará a: $${C= \frac{\epsilon_{0}.k.A}{d} \Longrightarrow C= \frac{\epsilon_{0}.(1+X_{e}).A}{d}}$$
Allí $X_{e}$ va a ser la constante dieléctrica del material entre las placas.

Para _ambas configuraciones_ se tomaron **5 medidas de capacidad**, se armó una tabla y se procedió con la realización de gráficos de dispersión para poder determinar la ${X_{e}}$


### _Mediciones y resultados_
Para la **_Primera configuración_** se utilizaron dos placas metálicas con separaciones de vidrios de distintos espesores.

Vidrio | d [m] | C [F]

1 placa C   | 5,90mm o sea  ${5,9.10^{-3}}$ m o sea 0,0059m| 0,326nF o sea ${3,26.10^{-10}}$ F

Con los datos obtenidos se realizó un gráfico con los valores de la capacidad _C [F]_ por los valores de $\frac{1}{d}$ (_d_ es la separación entre las placas metálicas), para así determinar la constante de susceptibilidad dieléctrica $X_{e}$



Comparando la formula de capacitancia (${C = \frac{\epsilon_{0}.k.A}{d}}$) con una recta (${y = m.x + b}$), se determinó  que ${\epsilon_{0}.k.A}$ es igual a la pendiente de la recta: $${m = \epsilon_{0}.k.A = 1,30.10^{-12}}$$

Para determinar el error de la pendiente se utilizo la siguiente fórmula: $${\Delta_{m} = \sqrt{\frac{m^{2}}{(N -2)}.(\frac{1}{R^{2}} - 1)}}$$
En dicha ecuación el valor de _N_ se corresponde con el número de mediciones realizadas. El gráfico nos proporcionó  el valor de la **pendiente _m_** y el **valor del coeficiente de determinación ${R^{2} = 0,986}$**, se reemplazaron ambos en la ecuación para encontrar el **error de la pendiente**: $${\Delta_{m} = \sqrt{\frac{(1,30.10^{-12})^{2}}{(5 -2)}.(\frac{1}{(0,986)} - 1)} = 8,94.10^{-14} = 0,089.10^{-12}}$$

Se llegó a la conclusión de que **la pendiente _m_ + su error** es igual a  $\Longrightarrow$   ${m = (1,30.10^{-12} \pm 0,089.10^{-12})}$

Utilizando la fórmula de la pendiente de la recta para despejar la **constante dieléctrica _k_** se llegó a que $${m = \epsilon_{0}.k.A \Longrightarrow k= \frac{m}{A.\epsilon_{0}}}$$

Entonces, reemplazando por:
1. Permitividad del vacío $\epsilon_{0}$: ${8,85.10^{-12} \frac{F}{m}}$
2. Area de las placas metálicas: ${20cm x 20cm = 0,2m x 0,2m= 0,04m^{2}}$
3. Pendiente: ${1,30.10^{-12}}$
$${k= \frac{1,30.10^{-12}}{0,04m^{2}.8,85.10^{-12}}= 3,672316384}$$

Se calculó el **error de la constante dieléctrica _k_**: $${\Delta_{m}= \Delta_{k}.\epsilon_{0}.A \Longrightarrow \Delta_{k}= \frac{\Delta_{m}}{\epsilon_{0}.A} =  \frac{0,089.10^{-12}}{8,85.10^{-12}.0,04} = 0,25}$$

Lo que nos deja como saldo final que **la constante dieléctrica _k_** es igual a $\Longrightarrow$ ${k = (4 \pm 0,3)}$. 
Utilizando la formula ${k = 1 + X_{e}}$ para calcular la susceptibilidad dieléctrica nos queda como resultado que: $${X_{e} = 3} \Longrightarrow {X_{e}= (3,0 \pm 0,3)}$$

```python
import matplotlib.pyplot as plt

#Defino los grosores de los vidrios y las capacidades como constantes, ya que el gráfico se encarga de relacionar los valores.
#Opté por esto en vez de crear los arrays en base a un diccionario u otras soluciones que generan muchas indirecciones para hacer lo mismo.
VIDRIO_B = 1/0.0038
VIDRIO_C = 1/0.0059
VIDRIO_BB = 1/0.0076
VIDRIO_BC = 1/0.0097
VIDRIO_CC = 1/0.0118

CAPACIDAD_B = 4.3e-10
CAPACIDAD_C = 3.26e-10
CAPACIDAD_BB = 2.74e-10
CAPACIDAD_BC = 2.3e-10
CAPACIDAD_CC = 2.02e-10

vidrios = [VIDRIO_B,VIDRIO_C,VIDRIO_BB,VIDRIO_BC,VIDRIO_CC]
capacidades_vidrio = [CAPACIDAD_B,CAPACIDAD_C,CAPACIDAD_BB,CAPACIDAD_BC,CAPACIDAD_CC]

plt.figure(figsize=(12,6))
plt.scatter(vidrios,capacidades_vidrio,c='purple')
plt.title('Láminas de vidrio')
plt.ylabel('C [F]')
plt.ylim(0, 5.0e-10)
plt.xlabel('1/d [1/m]')
plt.xticks(range(0,350,50))

#Tal vez hay una forma de que los valores del eje y se representen como 1.0e-10 en vez de tener la referencia por encima del eje.

plt.show()
```

Con los datos obtenidos se realizó un gráfico con los valores de la capacidad _C [F]_ por los valores de $\frac{1}{d}$ (_d_ es la separación entre las placas metálicas), para así determinar la constante de susceptibilidad dieléctrica $X_{e}$



1. **Pendiente** = ${7,74.10^{-13}}$
2. **valor del coeficiente de determinación** ${R^{2}}$ = ${0,987}$
3. **Numero de mediciones _N_** = ${5}$
4. **Permitividad del vacío** $\epsilon_{0}$ = ${8,85.10^{-12} \frac{F}{m}}$
5. **Area de las placas metálicas**: ${20cm x 20cm = 0,2m x 0,2m= 0,04m^{2}}$

Se calcula el **error de la pendiente $\Delta_{m}$**: $${\Delta_{m} = \sqrt{\frac{m^{2}}{(N -2)}.(\frac{1}{R^{2}} - 1)}}  \Longrightarrow \sqrt{\frac{(7,74.10^{-13})^{2}}{(5 -2)}.(\frac{1}{0,987} - 1)} = 5,12.10^{-14} = 0,51.10^{-13}$$

La **pendiente + su error** quedan definidos de la siguiente manera: ${m = (7,74.10^{-13} \pm 0,51.10^{-13})}$

Utilizando la fórmula de la pendiente de la recta para despejar la **constante dieléctrica _k_** se llegó a que $${m = \epsilon_{0}.k.A \Longrightarrow k= \frac{m}{A.\epsilon_{0}} = \frac{7,74.10^{-13}}{0,04m^{2}.8,85.10^{-12}} = 2,186440678}$$

Se calculó el **error de la constante dieléctrica _k_**: $${\Delta_{m}= \Delta_{k}.\epsilon_{0}.A \Longrightarrow \Delta_{k}= \frac{\Delta_{m}}{\epsilon_{0}.A} =  \frac{0,51.10^{-13}}{8,85.10^{-12}.0,04} = 0,1440677966}$$

Lo que nos deja como saldo final que **la constante dieléctrica _k_** es igual a $\Longrightarrow$ ${k = (2 \pm 0,1)}$. 
Utilizando la formula ${k = 1 + X_{e}}$ para calcular la susceptibilidad dieléctrica nos queda como resultado que: $${X_{e} = 1} \Longrightarrow {X_{e}= (1,0 \pm 0,1)}$$

```python
#También opté por definir constantes y luego armar los arrays, aunque el nombre de capacidad puede ser confuso ya que no está asociado al material de la placa.
ACRILICO_A = 1/0.0019
ACRILICO_D = 1/0.0037
ACRILICO_E = 1/0.0045
ACRILICO_ED = 1/0.0082
ACRILICO_EDA = 1/0.0101

CAPACIDAD_A = 4.55e-10
CAPACIDAD_D = 2.93e-10
CAPACIDAD_E = 2.3e-10
CAPACIDAD_ED = 1.47e-10
CAPACIDAD_EDA = 1.24e-10

acrilicos = [ACRILICO_A,ACRILICO_D,ACRILICO_E,ACRILICO_ED,ACRILICO_EDA]
capacidades_acrilico = [CAPACIDAD_A,CAPACIDAD_D,CAPACIDAD_E,CAPACIDAD_ED,CAPACIDAD_EDA]

plt.figure(figsize=(12,6))
plt.scatter(acrilicos,capacidades_acrilico,c='green')
plt.title('Láminas de acrílico')
plt.ylabel('C [F]')
plt.ylim(0, 5.0e-10)
plt.xlabel('1/d [1/m]')
plt.xticks(range(100,600,50))

#Tal vez hay una forma de que los valores del eje y se representen como 1.0e-10 en vez de tener la referencia por encima del eje.

plt.show()
```
